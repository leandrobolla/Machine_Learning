Dataframes COMBINACIONES: Concatenacion y Mezcla

```python
#CONCATENAR FILAS
df2 = pd.DataFrame({'NOMBRE':['Jose','Max'],'CARRERA':['Abogacia','Matematicas'],'EDAD': [23,26]}).set_index('NOMBRE')
df3 = pd.DataFrame({'NOMBRE':['Maria','John'],'CARRERA':['Fisica','Informatica'],'EDAD': [28,39]}).set_index('NOMBRE')
```

```python
df4 = pd.concat([df2,df3])
print(df4)
```

```python
#CONCATENAR COLUMNAS
df5 = pd.DataFrame({'AUTOS':['Nissan','Ford','Audi'],'COLOR':['Negro','Verde','Azul']}).set_index('AUTOS')
df6 = pd.DataFrame({'AUTOS':['Nissan','Ford','Audi'],'MODELO':['Tidda','Fiesta','TT']}).set_index('AUTOS')
```

```python
df7 = pd.concat([df5,df6], axis=1)
df7
```

MEZCLAR DATAFRAMES, SE HACEN DE A DOS UNICAMENTE, TIENE EN CUANTA LA MEZCLA DE INFORMACION EN COMUN Y SE COMBINAS POR UNA CLAVE.

```python
df8 = pd.DataFrame({'AUTOS':['Nissan','Ford','Audi'],'COLOR':['Negro','Verde','Azul']})
df9 = pd.DataFrame({'AUTOS':['Nissan','Ford','Audi'],'MODELO':['Tidda','Fiesta','TT']})
```

```python
df10 = pd.merge(df8,df9,on='AUTOS', how='outer')#'rigth','left','inner' #CON ON CREAMOS LA CLAVE y OUTER MEZCLA A NIVEL FILAS
```
